```python
from numpy_encode import *
from tqdm import tqdm

import music21
import glob
import numpy as np
```

```python
songs_path = []
for file in tqdm(glob.glob(".\\midi Melody Of The Night\\*.*"), desc="load path xml"):
        # chỉ chọn file đuôi .krn hoặc .mid
        if file[-3:] == "xml" or file[-3:] == "mid":
            songs_path.append(file)
```

```python
songs = []
for path in songs_path:
    songs.append(midi2npenc(midi_file = path))
```
